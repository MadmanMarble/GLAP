##**Assignment 1.3**: create at least two environments, each with different walls and with different sequence of visit locations.

```python
from agentEnv import Rob_body, Rob_env
from agentMiddle import Rob_middle_layer
from agentTop import Rob_top_layer, Plot_env
```

```python
# each paring is a tuple
# inner level () contain 2 points on the map. that forms the extrimities of one wall

env = Rob_env({((20,0),(30,20)), ((70,-5),(70,25))})
body = Rob_body(env)
middle = Rob_middle_layer(body)
top = Rob_top_layer(middle)
body.plotting = False
top.do({'visit':['o109','storage','o109','o103']})
middle.do({'go_to':(30,-10), 'timeout':200})
body.plotting = True
pl=Plot_env(body,top)
pl.plot_run()
```
