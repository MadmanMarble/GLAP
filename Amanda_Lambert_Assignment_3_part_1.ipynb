**1. Write a command that finds the mean of all numerical variables in titanic**

```python
# Determine which variables are numeric and which are not
print("Numeric variables:", df.select_dtypes(include=['number']).columns)
print("Object variables:", df.select_dtypes(include=['object']).columns)
```

```python
# Select only the numeric variables
numeric_df = df[['PassengerId', 'Class', 'SibSp', 'Parch', 'Fare']]
```

```python
# Find the mean of only the numeric variables
numeric_means = numeric_df.mean()
print("Means of  numeric variables: \n", numeric_means)
```

**2. Write a command that finds the minimum variables of all columns in the titanic**

```python
# 2. Write a command that finds the minimum values of all columns in titanic
numeric_mins = numeric_df.min()
print("Minimum of numeric variables:\n", numeric_mins)
```

**3. Write code hat uses the code from above to find the minimum age of all passengers in titanic**

```python
# Age column editing
print(df['Age'].unique())

# Change the wording of the age groups to all be just numbers
df['Age'] = df['Age'].replace({'Over 40': '40+', 'Under 18': '0 - 17'})
print(df['Age'].unique())
print("Minimum age of passengers:", df['Age'].min())
```
